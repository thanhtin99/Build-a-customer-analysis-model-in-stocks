```python
# import các thư viện cần thiết

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
```

```python
# Khởi tạo trình duyệt Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://www.hsx.vn/Modules/Rsde/RealtimeTable/LiveSecurity"
driver.get(url)

time.sleep(5)
```

```python
# Tìm bảng chứa dữ liệu cổ phiếu
table = driver.find_element(By.ID, "table-deal-stock")

html_content = table.get_attribute('outerHTML')
soup = BeautifulSoup(html_content, 'html.parser')

# Lấy tất cả các dòng dữ liệu (bỏ qua dòng tiêu đề)
rows = soup.find_all('tr')[1:]

data = []
for row in rows:
    if "main-header-row" not in row['class'] and "sub-header-row" not in row['class']:
        cols = row.find_all('td')
        if len(cols) >= 22:  
            stock_data = {
                "Mã CK": cols[0].text.strip(),
                "Đóng cửa": cols[1].text.strip(),
                "Trần": cols[2].text.strip(),
                "Sàn": cols[3].text.strip(),
                "Giá 3 Mua": cols[4].text.strip(),
                "KL 3 Mua": cols[5].text.strip(),
                "Giá 2 Mua": cols[6].text.strip(),
                "KL 2 Mua": cols[7].text.strip(),
                "Giá 1 Mua": cols[8].text.strip(),
                "KL 1 Mua": cols[9].text.strip(),
                "Giá khớp": cols[10].text.strip(),
                "Khối lượng khớp": cols[11].text.strip(),
                "Thay đổi giá": cols[12].text.strip(),
                "Tỷ lệ thay đổi": cols[13].text.strip(),
                "Giá 1 Bán": cols[14].text.strip(),
                "KL 1 Bán": cols[15].text.strip(),
                "Giá 2 Bán": cols[16].text.strip(),
                "KL 2 Bán": cols[17].text.strip(),
                "Giá 3 Bán": cols[18].text.strip(),
                "KL 3 Bán": cols[19].text.strip(),
                "TKL đã khớp": cols[23].text.strip(),  
                "KL NN mua": cols[25].text.strip(),  
                "KL NN bán": cols[26].text.strip() 
            }
            data.append(stock_data)
        else:
            print(f"Hàng dữ liệu không hợp lệ: {row}")
    else:
        print("Bỏ qua hàng tiêu đề hoặc tiêu đề phụ")


print(data)
```

```python
# Lưu dữ liệu thu thập vào file csv

import pandas as pd

df = pd.DataFrame(data)

df.to_csv('stock_data.csv', index=False)
```
